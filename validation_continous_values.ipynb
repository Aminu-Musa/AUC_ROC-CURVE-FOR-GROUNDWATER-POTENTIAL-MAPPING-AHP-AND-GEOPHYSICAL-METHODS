```python
import pandas as pd
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("predicted_and_validation_datasets_with_class2.csv")
```

```python
df.waterColumn.mean()
```

```python
df['True_label'] = df['waterColumn'].apply(lambda x: 1 if x >= 40.36 else 0)
```

```python
ahp_score = df.ahp
geophysics_score = df.geophysics
true_labels = df.True_label
```

```python
fpr1, tpr1, _= roc_curve(true_labels, ahp_score)
fpr2, tpr2, _= roc_curve(true_labels, geophysics_score)
```

```python
auc1 = roc_auc_score(true_labels, ahp_score)
auc2 = roc_auc_score(true_labels, geophysics_score)

print(f" AHP ROC score: {auc1:.2f}")
print(f" Geophysics ROC score: {auc2:.2f}")
```

<h3 style="color:yellow"> MODEL VALIDATION FOR AHP GWPZM </h3>

```python
# AHP ROC CURVE
plt.figure(figsize=(7,7))

# Plot each model with thinner lines and square markers
plt.plot(fpr1, tpr1, color='green', lw=1, label=f'AHP (AUC = {auc1:.2f})')

# Diagonal line (black dashed)
plt.plot([0, 1], [0, 1], color='black', linestyle='--', lw=1)

# Labels
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC Curves For AHP Method')

# Legend
plt.legend(loc='lower right')

# Remove grid
plt.grid(True)

# Tight layout, no margins
plt.margins(0)
plt.tight_layout(pad=0)

# Show plot
plt.show()
```

<h3 style="color:yellow"> MODEL VALIDATION FOR GEOPHYSICS GWPZM </h3>

```python
# GEOPHYSICS ROC CURVE
plt.figure(figsize=(7,7))

# Plot each model with thinner lines and square markers
plt.plot(fpr2, tpr2, color='red', lw=1, label=f'Geophysics (AUC = {auc2:.2f})')

# Diagonal line (black dashed)
plt.plot([0, 1], [0, 1], color='black', linestyle='--', lw=1)

# Labels
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC Curves For Geophysical Method')

# Legend
plt.legend(loc='lower right')

# Remove grid
plt.grid(True)

# Tight layout, no margins
plt.margins(0)
plt.tight_layout(pad=0)

# Show plot
plt.show()
```

<h3 style="color:yellow">COMBINED MODEL VALIDATION FOR AHP AND GEOPHYSICS GWPZM </h3>

```python
plt.figure(figsize=(7,7))

# Plot each model with thinner lines and square markers
plt.plot(fpr2, tpr2, color='red', lw=1,  label=f'Geophysics (AUC = {auc2:.2f})')
plt.plot(fpr1, tpr1, color='green', lw=1, label=f'AHP (AUC = {auc1:.2f})')

# Diagonal line (black dashed)
plt.plot([0, 1], [0, 1], color='black', linestyle='--', lw=1)

# Labels
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC Curves Comparison Between Geophysical and AHP Methods')

# Legend
plt.legend(loc='lower right')

# Remove grid
plt.grid(True)

# Tight layout, no margins
plt.margins(0)
plt.tight_layout(pad=0)

# Show plot
plt.show()
```
